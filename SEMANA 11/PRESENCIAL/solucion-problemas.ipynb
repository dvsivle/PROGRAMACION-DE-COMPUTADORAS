```python
#Problema 3:

class Fraccion(object):
    
    def __init__(self,numerador=0,denominador=1):
            self.numerador=numerador
            self.denominador=denominador
    
    #métodos decoradores:
    @property
    def numerador(self):
        return self.__numerador
    
    @numerador.setter
    def numerador(self,nuevoN):
        if(isinstance(nuevoN,int)):
            self.__numerador=nuevoN
        else:
            print("Error en el tipo de dato para el numerador!")

    @property
    def denominador(self):
        return self.__denominador
    
    @denominador.setter
    def denominador(self,nuevoD):
        if(isinstance(nuevoD,int)):
            if(nuevoD>0):
                self.__denominador=nuevoD
            else:
                print("El denominador no puede ser 0!")
        else:
            print("Error en el tipo de dato para el denominador!")
                        
    #sobrecargando al método privado __str__
    def __str__(self):
        return "{}/{}".format(self.numerador,self.denominador)
    
    #sobrecargando al método privado __add__
    def __add__(self,other):
        den = self.denominador*other.denominador
        num = self.numerador*other.denominador+other.numerador*self.denominador
        return Fraccion(num,den)
      
    #sobrecargando al método privado __mult__
    def __mul__(self,other):
        num = self.numerador*other.numerador
        den = self.denominador*other.denominador
        return Fraccion(num,den)
    
    #OPCIONAL
    def simplifica(self):   
        i=2
        while(1):
            menor=min(self.numerador,self.denominador)
            if(i<=menor):
                while(1):
                    if(self.numerador%i==0 and self.denominador%i==0):
                        self.numerador=int(self.numerador/i)
                        self.denominador=int(self.denominador/i)
                    else:
                        i+=1
                        break
            else:
                break
            
        return Fraccion(self.numerador,self.denominador)
```

```python
#se crea al objeto f de la clase Fraccion (se crea una fraccion)
#tambien se podria decir: se instancia un objeto de la clase Fraccion
f = Fraccion("888",0)

f = Fraccion(3,5)
print(f)
f.numerador="34"
print(f)
f.denominador="66"
print(f)
f.denominador=-5
print(f)

f = Fraccion(180,3600)
print(f)
fs = f.simplifica()
print("Fraccion simplificada: ",fs)
```
