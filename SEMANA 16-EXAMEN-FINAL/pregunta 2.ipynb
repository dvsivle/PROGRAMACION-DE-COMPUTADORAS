```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

while(1):
    a=int(input("ingrese cantidad e senoidales: "))
    if(0<a<=30):
        break
    else:
        print("El valor debe ser positivo y no mayor a 30")

t=np.linspace(0,0.01,1000)
cien=100
sumx=0.5
f=0

for i in range(a):
    f+=cien*np.sin(2*np.pi*cien*t)
    cien+=100
    sumx+=0.5
plt.plot(t,f)

plt.title("suma de senoides")
plt.ylabel("Amplitud")
plt.xlabel("tiempo[seg.]")
plt.grid()
```

```python
#pregunta 4

import numpy as np
import matplotlib.pyplot as plt



while(1):
    entrada=input('Ingrese cantidad de senoidales: ')
    if 0<int(entrada)<30:
        break
    else:
        entrada=0
        print('El valor debe ser positivo y no mayor a 30')
a=0
b=0
f=0
tiempo=np.linspace(0,0.01,10000)
for i in entrada:
    a=a+0.5
    b=b+100
    f=a*np.sin(2*np.pi*b*tiempo)
    
plt.plot(tiempo, f)
plt.title('Suma de senoidales')
plt.xlabel('Tiempo [seg.]')
plt.ylabel('Amplitud')
plt.grid()
```
