```python

from math import *

class Cuadratica(object):
    
    def __init__(self, a=0,b=0,c=0):
        self.a = a
        self.b = b
        self.c = c
    
    @property
    def a(self):
        return self.__a
    @a.setter
    def a(self,nuevoA):
        try:
            if(nuevoA>0):
                ax=nuevoA+1
                self.__a=nuevoA
            else:
                print("Error de valor")
        except TypeError:
            print("Error de tipo")
            
    @property
    def b(self):
        return self.__b
    @b.setter
    def b(self,nuevoB):       
        try:
            if(nuevoB>0):
                ax=nuevoB+1
                self.__b=nuevoB
            else:
                print("Error de valor")
        except TypeError:
            print("Error de tipo")
            
    @property
    def c(self):
        return self.__c
    @b.setter
    def c(self,nuevoC):
        try:
            if(nuevoC>0):
                ax=nuevoC+1
                self.__c=nuevoC
            else:
                print("Error de valor")
        except TypeError:
            print("Error de tipo")                  
    #mÃ©todos
    def discriminante(self):
        #dis=(self.a**2)-(4*self.a*(self.c))
        return (self.a**2)-(4*self.a*(self.c))
    
    def raices(self):
        x1 = (-self.b+self.discriminante()**(-2))/(2*self.a)
        x2 = (-self.b-self.discriminante()**(-2))/(2*self.a)
        return [x1,x2]
    
    def __add__(self,other):
        a1=self.a+other.a
        b1=self.b+other.b
        c1=self.c+other.c
        return Cuadratica(a1,b1,c1)
    def __str__(self):
        return "{:.2f}x^2 + {:.2f}x +  {:.2f} = 0".format(self.a,self.b,self.c)
```

```python
ec1= Cuadratica(2,4,5)
ec2= Cuadratica(4,2,6)

print("raices de ec1: {:.3f},{:.3f}".format(ec1.raices()[0],ec1.raices()[1]))
print("raices de ec2: {:.3f},{:.3f}".format(ec2.raices()[0],ec2.raices()[1]))
print("ec1 + ec2=",ec1+ec2)
ec1.a='2'
ec1.b=-3
ec1.c=0
```

```python
#pregunta 2:


#creacion de la clase
import math
class Caudratica(object):
    def _init_(self, a=None, b=None, c=None):
        self.a=a
        self.b=b
        self.c=c
        
    #decorativos:
    @property
    def a(self):
        return self.__a
    @a.setter
    def a(self, Na):
        if (isinstance(Na, int) or isinstance(Na, float) and (Na>1)):
            self.__a = Na
        else:
            print('datos incorrectos')
            
    @property
    def b(self):
        return self.__b
    @b.setter
    def b(self, Nb):
        if (isinstance(Nb, int) or isinstance(Nb, float) and (Nb>1)):
            self.__b=Nb
        else:
            print('Datos incorrectos')
    
    @property
    def c(self):
        return self.__c
    @c.setter
    def c(self, Nc):
        if (isinstance(Nc, int) or (isinstance(Nc, float))):
            self.__c=Nc
        else:
            print('Datos incorrectos')
        
            
    #creacionde los metodos:
    def discriminante(self):
        discri = self.b**2-4*self.a*self.c 
        return discri
    
    def raices(self):
        num=self.discriminante()
        if num>0:
            raiz1=(-self.b+math.sqrt(num))/2*self.a
            raiz2=(-self.b-math.sqrt(num))/2*self.a
            return [raiz1, raiz2]
        else:
            pass
            
    #cargando los metodos:
    def _str_(self):
        return '{:.2f}X^2 + {:.2f}X + {:.2f} = 0'.format(self.a, self.b, self.c)
    
    def _add_(self, other):
        A=self.a+other.a
        B=self.b+other.b
        C=self.c+other.c
        return Caudratica(A, B, C)
```
