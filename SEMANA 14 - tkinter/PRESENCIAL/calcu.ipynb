```python
from tkinter import *

root = Tk()
#variavles generales del root
colorFondo="red"
colorLetra = "#FFF"


def suma():
    x= int(Lresul["text"])+1
    Lresul["text"]="{}".format(x)

def resta():
    x= int(Lresul["text"])
    Lresul["text"]="{}".format(x-1)


#root.geometry("300x60")
root.title("Calculadora")
root.configure(background = colorFondo) #color de fondo


Bmenos= Button(master=root, text="-", bg="#009", fg="#FFF", width=8, height=2, font=("Helvetica", 15),command=resta)
Bmenos.pack(side=LEFT)

Lresul = Label (master=root, text="0", bg="red",  fg="#FFF", width=8, height=2, font=("Helvetica", 15))
Lresul.pack(side=LEFT)

Bmas  = Button(master=root, text="+", bg="#009", fg="#FFF", width=8, height=2, font=("Helvetica", 15),
             command=suma)
Bmas.pack(side=LEFT)

root.mainloop()
```
