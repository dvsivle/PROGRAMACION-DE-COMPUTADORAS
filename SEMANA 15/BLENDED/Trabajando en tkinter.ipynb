```python
def evalua(n1,n2,n3):
    if n2%n1==0 or n3%n1==0:
        if n2%n1==0 and n3%n1==0:
            #return [n1,n2,n3]
            print("\n{} y {} son múltiplos de {}.".format(n3,n2,n1))
        elif n2%n1==0:
            #return [n1,n2]
            print("\n{} es múltiplo de {}.".format(n2,n1))
        else:
            #return [n1,n3]
            print("\n{} es múltiplo de {}.".format(n3,n1))
    if n3%n2==0 :
        #return [n2,n3]
        print("\n{} es múltiplo de {}.".format(n3,n2))
    else:
        print("\nNingún número es múltiplo de otro.")
    
print("DETECTOR DE MULTIPLOS")
print("Escriba tres numeros enteros crecientes.")
print("Le indicare si alguno es multiplo de otro")

while(1):
    n1=int(input("Primer número: "))
    n2=int(input("Segundo número: "))
    
    while (n2<=n1):
        print("\nPor favor, escriba números crecientes")
        n2=int(input("Segundo número: "))
        
    n3=int(input("Tercer número: "))
    while (n3<=n2):
        print("\nPor favor, escriba números crecientes")
        n3=int(input("Tercer número: "))
    break

evalua(n1,n2,n3)
```
