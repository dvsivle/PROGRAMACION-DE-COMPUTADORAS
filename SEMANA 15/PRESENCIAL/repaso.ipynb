```python
# Ajuste de estilo (otra forma de configurar el color y estilo de línea)
import numpy as np
import matplotlib.pyplot as plt

P=10**6  
dom = np.linspace(0, 2*np.pi, 100) #dominio

mercurio = (0.206*269.2*P)/(1-0.206*np.cos(dom))
venus = (0.00677*15913*P)/(1-0.00677*np.cos(dom))
tierra = (0.0167*8964*P)/(1-0.0167*np.cos(dom))
marte = (0.0934*2421*P)/(1-0.0934*np.cos(dom))

plt.plot(dom, mercurio,label="Mercurio")
plt.plot(dom, venus,label="Venus")
plt.plot(dom, tierra,label="Tierra")
plt.plot(dom, marte,label="Marte")

plt.title("Orbita de 4 planetas")
plt.legend(loc='lower left')
plt.grid()
plt.show()
```
