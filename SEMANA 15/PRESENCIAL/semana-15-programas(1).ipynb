```python
#PROBLEMA 1:
import numpy as np
import matplotlib.pyplot as plt

#creando el dominio:
dom = np.linspace(0, 2*np.pi, 100) #dominio

#creando los rangos:
f=10**6 
mercurio = (0.206*269.2*f)/(1-0.206*np.cos(dom))
venus = (0.00677*15913*f)/(1-0.00677*np.cos(dom))
tierra = (0.0167*8964*f)/(1-0.0167*np.cos(dom))
marte = (0.0934*2421*f)/(1-0.0934*np.cos(dom))

plt.plot(dom, mercurio)
plt.plot(dom, venus)
plt.plot(dom, tierra)
plt.plot(dom, marte)

#tambien puede ser asi:
#plt.plot(dom, mercurio,dom, venus,dom, tierra,dom, marte)

plt.title("Orbita de 4 planetas")
plt.grid()

#agregando una leyenda
plt.legend(['Mercurio','Venus','Tierra','Marte'],loc='upper right')
#plt.show()
```

```python
#PROBLEMA 2:

import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0, 4, 100) #dominio

distancia=np.sqrt((14-8*t)**2+(25-16*t)**2)  #rango

plt.plot(t,distancia)
plt.xticks([0,1,2,3,4],['7AM','8AM','9AM','10AM','11AM'])
plt.xlabel("Tiempo [horas]")
plt.ylabel("distancia [millas]")
plt.title("distancia entre los Barcos")
plt.grid()

#opcional 1:
mx=t[distancia==np.min(distancia)]
plt.plot(mx,np.min(distancia),'ko')
plt.text(mx-0.9,np.max(np.min(distancia))-1,"[{:.2f},{:.2f}]".format(float(mx),float(np.min(distancia))),bbox=dict(facecolor='magenta', alpha=0.2))
plt.show()

#opcional 2:
#formar una tabla:
print("\ntiempo\tdistancia")
#for tt,dd in zip(t,distancia):
#    print("{:.3f}\t{:.3f}".format(tt,dd))


#opcional 3:
dm = np.min(distancia)
print("\nLa distancia minima entre los barcos: {:.3f} millas".format(dm))
#a que hora los barcos se encontran a la minima distancia:
tm = t[distancia == dm]
print("\nSe encontraran a la minima distancia a las {:.3f} AM".format(7+tm[0]))
```

```python
#Problema 3:

from math import *

class Osc555(object):
    
    def __init__(self, R1=1,R2=1,C=1):
        self.R1 = R1
        self.R2 = R2
        self.C = C
    
    @property
    def R1(self):
        return self.__R1
    @R1.setter
    def R1(self,nuevoR1):
        try:
            if(nuevoR1>0):
                self.__R1=nuevoR1
            else:
                print("Error de valor")
        except TypeError:
            print("Error de tipo")
            
    @property
    def R2(self):
        return self.__R2
    @R2.setter
    def R2(self,nuevoR2):       
        try:
            if(nuevoR2>0):
                self.__R2=nuevoR2
            else:
                print("Error de valor")
        except TypeError:
            print("Error de tipo")
            
    @property
    def C(self):
        return self.__C
    @C.setter
    def C(self,nuevoC):
        try:
            if(nuevoC>0):
                self.__C=nuevoC
            else:
                print("Error de valor")
        except TypeError:
            print("Error de tipo")                  
    #mtodos
    
    def frecuencia(self):
        f=1/(log(2)*self.C*(self.R1+2*self.R2))
        return f
    def periodo(self):
        T=1/self.frecuencia()
        return T
    def duty_cycle(self):
        D=(log(2)*(self.R1+self.R2)*self.C)/self.periodo()
        return D
    
    def __str__(self):
        return "OSC555[R1={:.2f}, R2={:.2f}. C={:.2f}]".format(self.R1,self.R2,self.C)
```

```python
#comprobando la clase:
obj = Osc555(100,200,0.1)

print(obj)

obj.R1 = "34"
obj.R2 = 0
obj.C = -4

obj.R1 = 34
obj.R2 = 45
obj.C = 4

print(obj)
print("Frecuencia: {:.2f} Hz".format(obj.frecuencia()))
print("Periodo: {:.2f} Seg.".format(obj.periodo()))
print("Duty Cycle: {:.2f}".format(obj.duty_cycle()))
```
