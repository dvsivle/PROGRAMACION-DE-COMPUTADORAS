### Información preliminar
En el sistema operativo UNIX existe el comando "banner", que convierte una cadena de caracteres en un formato de letra grande, conformado de caracteres "#" en arreglos de 7x7 cada uno. Esta instrucción permite mostrar un mensaje en forma grande en los terminales.

Para el caso de este trabajo se debe de mostrar un "banner" pero con la hora actual del sistema, obtenida desde el reloj de la PC.

La hora del sistema se puede obtener gracias a la librería `time`:

```python
from time import strftime

time_str = strftime("%H:%M:%S")
print(time_str)
```

La instrucción `strftime` retorna una cadena con la hora actual del sistema. La cadena de configuración especifica los datos y el formato a retornar (en este caso, retorna la hora, dos puntos, el minuto, dos puntos y el segundo.

Su tarea consiste en convertir esta cadena en un banner de UNIX, de forma tal que se muestre la hora actual en caracteres de 7x7 de "#", lo que se verá como un arreglo matricial de 7 filas por 56 columnas. 56 porque en lugar de ":" se deberá mostrar un espacio en blanco (ANALICE ESE CASO)

**2.** La cadena al ser iterable sera evaluada caracter por caracter, de acuerdo a ello se aramará el "**baner**" (todo este proceso se desarrolla dentro de una función):  
  
>* Dentro de la función "**banner(hora)**" se crea una lista ( *BANNER* ), cuyos elemntos son 7 listas vacías, las cuales seran poblados de acuerdo al argumento recibido, que es un cadena que contiene la hora actual.  
  
>* Para poblar las sub-listas vacías de "*BANNER*" se realiza en un **for** donde se evaluará cada caracter de la cadena recbida. De acuerdo a ello se invoca a otra función "**agregar(L_agregar,L_matriz)**". Los datos de las sub-listas de "**L_agregar**" (contiene la lista que representa al caracter de la cadena) se agregan al a las bulistas de **L_matriz** (donde ) y es retornada para actualizar la lista "*BANNER*"  
  
>* Después de poblar " *BANNER* " con todos los caracteres de la cadena se retorna esta lista para poder imprimir los y mostras en la pantalla de la siguiente manera:

```python
#coloque aquí el nombre de los integrantes del grupo:
#1.- Huallanca Escalera, Harold 
#2.- Salcedo Rodríguez, Joan
#3.- De la torre Velarde, Elvis 

# Crear una lista que estará conformado por 7  sublistas 
#la cual sera representado cada caracter numerico  

numero_0 = [[' ',' ','#','#','#',' ',' ',' '],
            [' ','#',' ',' ',' ','#',' ',' '],
            ['#',' ','#',' ',' ',' ','#',' '],
            ['#',' ',' ','#',' ',' ','#',' '],
            ['#',' ',' ',' ','#',' ','#',' '],
            [' ','#',' ',' ',' ','#',' ',' '],
            [' ',' ','#','#','#',' ',' ',' ']]

numero_1 = [[' ',' ',' ','#',' ',' ',' ',' '],
            [' ',' ','#','#',' ',' ',' ',' '],
            [' ','#',' ','#',' ',' ',' ',' '],
            [' ',' ',' ','#',' ',' ',' ',' '],
            [' ',' ',' ','#',' ',' ',' ',' '],
            [' ',' ',' ','#',' ',' ',' ',' '],
            [' ','#','#','#','#','#',' ',' ']]

numero_2 = [[' ','#','#','#','#','#',' ',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            [' ',' ',' ',' ',' ',' ','#',' '],
            [' ','#','#','#','#','#',' ',' '],
            ['#',' ',' ',' ',' ',' ',' ',' '],
            ['#',' ',' ',' ',' ',' ',' ',' '],
            ['#','#','#','#','#','#','#',' ']]

numero_3 = [[' ','#','#','#','#','#',' ',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            [' ',' ',' ',' ',' ',' ','#',' '],
            [' ','#','#','#','#','#',' ',' '],
            [' ',' ',' ',' ',' ',' ','#',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            [' ','#','#','#','#','#',' ',' ']]

numero_4 = [['#',' ',' ',' ',' ',' ',' ',' '],
            ['#',' ',' ',' ',' ','#',' ',' '],
            ['#',' ',' ',' ',' ','#',' ',' '],
            ['#','#','#','#','#','#','#',' '],
            [' ',' ',' ',' ',' ','#',' ',' '],
            [' ',' ',' ',' ',' ','#',' ',' '],
            [' ',' ',' ',' ',' ','#',' ',' ']]

numero_5 = [['#','#','#','#','#','#','#',' '],
            ['#',' ',' ',' ',' ',' ',' ',' '],
            ['#',' ',' ',' ',' ',' ',' ',' '],
            [' ','#','#','#','#','#',' ',' '],
            [' ',' ',' ',' ',' ',' ','#',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            [' ','#','#','#','#','#',' ',' ']]

numero_6 = [[' ','#','#','#','#','#','#',' '],
            ['#',' ',' ',' ',' ',' ',' ',' '],
            ['#',' ',' ',' ',' ',' ',' ',' '],
            ['#','#','#','#','#','#',' ',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            [' ','#','#','#','#','#',' ',' ']]

numero_7 = [['#','#','#','#','#','#','#',' '],
            ['#',' ',' ',' ',' ','#',' ',' '],
            [' ',' ',' ',' ','#',' ',' ',' '],
            [' ',' ',' ','#',' ',' ',' ',' '],
            [' ',' ','#',' ',' ',' ',' ',' '],
            [' ',' ','#',' ',' ',' ',' ',' '],
            [' ',' ','#',' ',' ',' ',' ',' '],]

numero_8 = [[' ','#','#','#','#','#',' ',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            [' ','#','#','#','#','#',' ',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            [' ','#','#','#','#','#',' ',' ']]

numero_9 = [[' ','#','#','#','#','#',' ',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            [' ','#','#','#','#','#','#',' '],
            [' ',' ',' ',' ',' ',' ','#',' '],
            ['#',' ',' ',' ',' ',' ','#',' '],
            [' ','#','#','#','#','#',' ',' ']]


espacio = [[' ']*4]*7 


#FUNCIONES A UTILIZAR EN EL PROGRAMA

def agregar(L_agregar,L_matriz):
    
    for i in range(len(L_matriz)):   
        L_matriz[i].extend(L_agregar[i]) # extend() agrega toda una lista al final
            
    return L_matriz


def banner(hora):
    BANNER=[[],[],[],[],[],[],[]] #se crea una lista principal con 7 sublistas vacias
    
    for car in hora:
        if car == '0':
            BANNER=agregar(numero_0,BANNER)
        
        elif car =='1':
            BANNER=agregar(numero_1,BANNER)
            
        elif car =='2':
            BANNER=agregar(numero_2,BANNER)
            
        elif car =='3':
            BANNER=agregar(numero_3,BANNER)
            
        elif car =='4':
            BANNER=agregar(numero_4,BANNER)
            
        elif car =='5':
            BANNER=agregar(numero_5,BANNER)
            
        elif car =='6':
            BANNER=agregar(numero_6,BANNER)
            
        elif car =='7':
            BANNER=agregar(numero_7,BANNER)
            
        elif car =='8':
            BANNER=agregar(numero_8,BANNER)
        
        elif car =='9':
            BANNER=agregar(numero_9,BANNER)
            
        else:
            BANNER=agregar(espacio,BANNER)
        
    return BANNER  



#capturar la hora actual y almacenar en una cadena
from time import strftime

time_str = strftime("%H:%M:%S")
BANNER=banner(time_str)

print('Hora actual:',time_str)
print('Dimensión del banner  {}x{}'.format(len(BANNER),len(BANNER[0])))
print()

for i in range(len(BANNER)):  #para i:0,1,2,3,4,5,6 =># sub-listas
    print('\t\t\t',end='')
    for j in range(len(BANNER[i])): #para j:0,1,2..,len(BANNER[i])-1 =>posiciones de los elementos de la sub-lista
        print(BANNER[i][j],end="")
    print()
        
```
