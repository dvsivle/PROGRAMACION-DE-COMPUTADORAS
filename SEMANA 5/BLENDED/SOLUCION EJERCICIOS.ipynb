```python
from math import*
from random import*
from random import sample


def existenciaTriangulo(a,b,c):
    if (abs(a-b)<c and c<(a+b)):
        return True
    else:
        return False
    
def calculaArea(a,b,c):
    s=(a+b+c)/2
    área=sqrt(s*(s-a)*(s-b)*(s-c))
    return área

# MAIN

triangulos=[]
conjunto=[]
while(len(triangulos)<15):
    L=sample(range(5,15),3)
    if(existenciaTriangulo(L[0],L[1],L[2])):
        s=set(L)
        aux=0
        for a in conjunto:
            if(s==a):
                aux=1
                
        if(aux==0):
            triangulos.append(L)
            conjunto.append(s)
cont=1
for a in triangulos:
    print("Triángulo",cont)
    print("Lados: {},{},{}".format(a[0],a[1],a[2]))
    print("Área: {:.1f}".format(calculaArea(a[0],a[1],a[2])))
    print()
    cont+=1
    
   
```

# PREGUNTA 6

```python
from math import*
F=lambda x:(x**2 +2)/(2*x + 1)

def factorial(num):
    fact=1
    for i in range(1,num+1):
        fact=fact*i
    return fact

def calcularSuma(n):
    sumatoria=0
    for i in range(1,n+1):
        sumatoria=sumatoria+factorial(i)/(pow(F(i),2)+1)
        #print(factorial(i))
        #print(F(i))
        #print()
    return sumatoria

while(True):
    n=int(input("Ingrese la cantidad de terminos de la serie: "))
    if(n>0 and n<=15):
        break
    else:
        print("Error, ingrese nuevamente\n")
        
print("La sumatoria de la serie es: {}".format(calcularSuma(n)))
```

```python
from math import*
from random import* 


def promedio(lista):
    acum=0
    for i in lista:
        acum= acum + i 
    return acum/len(lista)

def varianza(lista):
    var=0
    prom=promedio(lista)
    for a in lista:
        var=var + (a-prom)**2
        
    return var/len(lista)



#main():
while(True):
    N=int(input("Ingrese un numero positivo: "))
    if(N>0):
        break
    else:
        print("Error, vuelva a ingresar")
#desarrollo()



L=[]
for a in range(N):
    alea=uniform(0,1)
    L.append(alea)
    
print()
print('Elementos de la lista:')
for i in L:
    print('{:.2f}'.format(i),end=('  '))
    

print('\n\nVarianza de la lista: {:.3f}'.format(varianza(L)))
```

```python
from math import*
from random import*
from random import sample


def existenciaTriangulo(a,b,c):
    if (abs(a-b)<c and c<(a+b)):
        return True
    else:
        return False
    
def calculaArea(a,b,c):
    s=(a+b+c)/2
    área=sqrt(s*(s-a)*(s-b)*(s-c))
    return área

def buscarEnlista(L,M):
    M.sort()
    for i in L:
        a=list(i)
        a.sort()
        if (M==a):
            return False
    return True

# MAIN

triangulos=[]

while(len(triangulos)<15):
    L=sample(range(5,15),3)
    
    if(existenciaTriangulo(L[0],L[1],L[2])):
         if(buscarEnlista(triangulos,L)):
            triangulos.append(L)
            
cont=1
for a in triangulos:
    print("Triángulo",cont)
    print("Lados: {},{},{}".format(a[0],a[1],a[2]))
    print("Área: {:.1f}".format(calculaArea(a[0],a[1],a[2])))
    print()
    cont+=1
    
        
```
