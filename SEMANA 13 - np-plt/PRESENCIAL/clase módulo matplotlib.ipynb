```python

import numpy as np
import matplotlib.pyplot as plt

n=int(input("ingrese el numero de ciclos: "))
A= 2  
f= 5
t = np.linspace(0, n/f, 1000)  
y1 = A * np.sin(2 * np.pi * f * t)

plt.plot(t, y1, color='r')

plt.title("Seal senoidal")
plt.xlabel("Tiempo [seg]")
plt.ylabel("Voltios [v]")

plt.grid()
```

```python

import numpy as np
import matplotlib.pyplot as plt

A= 2  
f= 5
T=1/f
t=np.linspace(0,T,1000)

y1= A*np.sin(2 * np.pi * f * t)
y2= 2*A*np.sin(2 * np.pi * f * t)
y3= (A/2)*np.sin(2 * np.pi * f * t)

plt.plot(t,y1,'--r')
plt.plot(t,y2,':g')
plt.plot(t,y3,'b')

plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

seal = np.random.randint(-10,11,20)


plt.plot(seal)
plt.grid()
cont=0
for i in range(1,seal.size-1):
    if(seal[i]>seal[i+1] and seal[i]>seal[i-1] and seal[i]>0):
        cont=cont+1
        plt.plot(i,seal[i],'or')
      
print("Hay ",cont," picos") 
plt.show()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy import special

def drumhead_height(n, k, distance, angle, t):
    kth_zero = special.jn_zeros(n, k)[-1]
    return np.cos(t) * np.cos(n*angle) * special.jn(n, distance*kth_zero)

np.r_[np.array([1,2,3]), 0, 0, np.array([4,5,6])]
theta = np.r_[0:2*np.pi:50j]
radius = np.r_[0:1:50j]
x = np.array([r * np.cos(theta) for r in radius])
y = np.array([r * np.sin(theta) for r in radius])
z = np.array([drumhead_height(1, 1, r, theta, 0.5) for r in radius])

### graficando

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.jet)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
