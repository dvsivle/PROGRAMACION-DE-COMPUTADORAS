```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#graficando una lista en función de sus índices:

L = [3,7,1]

#plt.plot([3,7,1])  #tambien se puede
plt.plot(L)
```

```python
#graficando los elementos de una lista en funcion de los elementos de otra:

plt.plot([1,2,3,4],[8,1,0,5]) #las 2 listas deben ser del mismo tamaño
```

```python
#graficando los elementos de una lista en función de los elementos de otra:

plt.plot([2,1,4,3],[8,1,0,5])  #OJO: la lista de abcisa esta desordenada
```

```python
#funciones adicionales para el formateo de la gráfica:

plt.plot([1,2,3,4],[8,1,0,5])
plt.title("Grafica de prueba")  #coloca título a la gráfica
plt.xlabel("eje x")  #etiqueta texto en el eje de abcisa
plt.ylabel("eje y")  #etiqueta texto en el eje de ordenada
plt.grid() #activa una cuadrícula en la gráfica
plt.axis([-1,6,-1,10]) #configura el rango de visualizacipon de los ejes x y
```

```python
#utilizando una cadena como argumento adicional al plot para formatear la gráfica:

plt.plot([1,2,3,4],[8,1,0,5],"--sb",linewidth=5,markersize=15,markerfacecolor='g')
```

```python
#graficando la funcion F(x)=x^2 para el dominio [0,4]

x = np.arange(0,5,0.2)
y = x**2

plt.plot(x,y)
```

```python
#graficando la funcion F(x)= sen(pi.x)/(pi.x) para el dominio [-10,10]

x = np.linspace(-10,10,500)
y = np.sin(np.pi*x)/(np.pi*x)

plt.plot(x,y)
```

```python
#requerimiento 5:
#f(t) = Asen(2.pi.f.t)

A=2
f=5

n = int(input("Ingrese número de períodos: "))
T =1/f

#n=1    t:[0,T]
#n=2    t:[0:2T]
#n=3    t:[0:3T]
#n      t:[0:nT]

t = np.linspace(0,n*T,1000) #arreglo de abcisa
y = A*np.sin(2*np.pi*f*t)

plt.plot(t,y,'r')
plt.title("Señal senoidal")
plt.xlabel("tiempo [Seg.]")
plt.ylabel("Voltios [V]")
plt.grid()
```

```python
#colocando 2 o mas gráficas en una misma ventana

A=2
f=5
T=1/f

t = np.linspace(0,T,1000) #arreglo de abcisa

#se crean 3 ordenadas (3 señales)
y1 = A*np.sin(2*np.pi*f*t)
y2 = 2*A*np.sin(2*np.pi*f*t)
y3 = (A/2)*np.sin(2*np.pi*f*t)

#plt.plot(t,y1,'--g')
#plt.plot(t,y2,':b')
#plt.plot(t,y3,'-.r')

#de otra forma:
plt.plot(t,y1,'--g',t,y2,':b',t,y3,'-.r')
```

```python
#requerimiento 6:

n = np.linspace(0,1,15)

v = np.random.rand(15)

plt.plot(n,v)

#hallando el valor máximo de v:
mv = np.max(v)

#hallando el valor n correspondiente al valor maximo mv
mx = n[v==mv]

plt.plot(mx,mv,'or')
```
