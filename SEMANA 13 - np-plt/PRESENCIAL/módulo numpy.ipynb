```python
import numpy as np
```

```python
#creando arreglos (arrays)

#crear un arreglo en base a una lista:
arr1 = np.array([5,7,8,10,-1])
print(type(arr1))
print("\n",arr1)

#crear un arreglo en base al método arange():
arr2 = np.arange(10)  #Se crea un arreglo que empieza en 0, termina en 9, con saltos de 1 en 1
print("\n",arr2)
arr3 = np.arange(3,11,2) #Se crea un arreglo que empieza en 3, termina en 3, con saltos de 2 en 2
print("\n",arr3)
arr4 = np.arange(0,11,0.2) #Se crea un arreglo que empieza en 0, termina en 10, 
                           #con saltos de 0.2 en 0.2
print("\n",arr4)

#crear un arreglo en base al método linspace():
arr5 = np.linspace(0,1,10) #crea un arreglo que empieza en 0, termina en 1 y de tamaño 10
print("\n",arr5)
arr6 = np.linspace(3,20,20) #crea un arreglo que empieza en 3, termina en 20 y de tamaño 20
print("\n",arr6)

#crear un arreglo en base al módulo random de numpy:
arr7 = np.random.rand(10) #arreglo de tamaño 10 con valores aleatorios reales en el rango [0,1>
print("\n",arr7)
arr8 = np.random.randint(1,7,15) #arreglo de tamaño 15 con valores aleatorios enteros en el 
                                 #rango [1,7>
print("\n",arr8)
arr9 = np.random.uniform(5,11,10) #arreglo de tamaño 10 con valores aleatorios reales en el 
                                  #rango [5,11>
print("\n",arr9)
```

```python
#Que podemos hacer con un arreglo:

arr1 = np.array([5,7,8,10,-1])
print("Tamaño del arreglo 1: ",len(arr1))
print("Cantidad de elementos del arreglo 1: ",arr1.size)

#los elementos de un arreglo al igual que en una lista estan indexados:
arr2 = np.arange(10)
print("\nAccede al primer elemento del arreglo 2: ",arr2[0])
print("Accede al ultimo elemento del arreglo 2: ",arr2[-1])
print("Accede a un tramo del arreglo 2: ",arr2[2:7])

#desde el punto de vista de modificación, el arreglo si es mutable
arr3 = np.arange(3,11,2) 
arr3[1]=100  #modifica el elemento de la posicion 1 por 100
print("\n",arr3)

#desde el punto de vista de insercion el arreglo es inmutable
arr4 = np.arange(0,11,0.2)
arr4 = np.insert(arr4,0,100) #tendría que actualizar
print("\n",arr4)

#desde el punto de vista de la eliminación el arreglo es inmutable
arr5 = np.linspace(0,1,10)
arr5 = np.delete(arr5,1) #tendría que actualizar
print("\n",arr5)
```

```python
#operaciones matemáticas aplicadas a un arreglo:

ltemp = [56, 12.5, 19.5, 29, 100, 30]  #lista de grados centigrados

#usando for y append():
lF = []
for el in ltemp:
    lF.append(el*9/5+32)

print(lF)

#usando mapeo:
lF = list(map(lambda x:x*9/5+32,ltemp))
print("\n",lF)

#usando arreglos:
aTemp = np.array([56, 12.5, 19.5, 29, 100, 30])  #creando el arreglo de temperaturas
aF = aTemp*9/5+32  #creando el arreglo de grados F

print("\n",aF)

#un arreglo es iterable:
print()
for el in aF:
    print(el)
```

```python
#requerimiento 1:

R = np.array([10,20,3.5,100,200])*1000  #se crea el arreglo de resistencias

V = np.array([120,80,110,200,350])  #se crea el arreglo de voltajes

#creando al arreglo corrientes:
I = V/R

#creando al arreglo de potencias:
P = V*I

print("R(ohmios)\tV(voltios)\tI(amp.)\t\tP(watts)")
for r,v,i,p in zip(R,V,I,P):
    print("{:7}k\t{:6}\t\t{:5.2f}\t\t{:6.2f}".format(r/1000,v,i,p))
```

```python
#Aplicando otras operaciones matemáticas a un arreglo:

A = np.random.randint(1,11,10)
print(A)

#calcular el seno de cada valor de A
sA = np.sin(A)
print("\n",sA)

#calcuar la exponencial de cada valor de A
eA = np.exp(A)
print("\n",eA)

#elevar al cuadrado cada valor de A
cA = A**2
print("\n",cA)

#multiplicar cada valor de A por 5
a5 = A*5
print("\n",a5)

#sumar 3 a cada valor de A 
a3 = A+3
print("\n",a3)

print("\nLa suma de los elementos de A es: ",np.sum(A))
print("El promedio de los elementos de A es: ",np.mean(A))
print("El maximo valor de los elementos de A es: ",np.max(A))
print("El minimo valor de los elementos de A es: ",np.min(A))

az = np.zeros(10)  #arreglo de 10 ceros
print("\n",az)

au = np.ones(10)  #arreglo de 10 unos
print("\n",au)

#arreglo de 20 valores 10
a10 = np.ones(10)*10
print("\n",a10)

arr = np.array([1,2,3,4,5,6,7])
print("\n",arr)
print("La suma acumulada de arreglo es: ",np.cumsum(arr))
```

```python
#requerimiento 2:

i= np.arange(0,51,dtype=float)
au = (-3)**(-i)/(2*i+1)
cs = np.cumsum(au)*np.sqrt(12)

for k,r in enumerate(cs):
    print("Valor de pi para n={} :{:.15f}".format(k+1,r))
```

```python
#indexamiento booleano:

A = np.random.randint(1,100,24)
print(A)

ar = A%2==0  #ar es un arreglo booleano (solo contiene True o False)
print("\n",ar)
pares = A[ar]  #accederá a los elementos de A donde el indice del [] sea True
print(pares)

#simplificando:

#crea un arreglo con todos los pares de A:
pares = A[A%2==0]
print("\n",pares)

#crea un arreglo con todos los multiplos de 7 de A:
m7 = A[A%7==0]
print("\n",m7)

#crea un arrego con todos los valores de A mayores a 60
m60 = A[A>60]
print("\n",m60)

#almacena en un arreglo todos los valores 50 de A
m50 = A[A==50]
print("\n",m50)

#crea un arreglo con todos los valores de A que esten en el rango [50, 80]
arr = A[A>=50]
arr = arr[arr<=80]
print("\n",arr)

print("\n",any(A>50)) #hay algún elemento de A mayor a 50??
print("\n",all(A<200))  #todos los elementos de A son menores a 200?
```

```python
#requeremiento 3:

#se crea el arreglo de nombres
nombres = np.array(["Juan","Maite","Laura","Richard","Carmen","Enrique"])

#creando el arreglo de notas:
notas = np.random.randint(0,21,6)

#mostrando la tabla:
for nomb,nota in zip(nombres,notas):
    print("{:8}:{:5}".format(nomb,nota))

#calculando el promedio:
prom = np.mean(notas)

nS = nombres[notas>prom]  #accede a los nombres cuyas notas pasaron el promedio
                          #y almacenalos en un arreglo

print("\nAlumnos cuyas notas estan por encima del promedio {:.2f}".format(prom))
for nomb in nS:
    print(nomb)
```
